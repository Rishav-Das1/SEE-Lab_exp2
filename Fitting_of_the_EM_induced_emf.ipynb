```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the form of the function you want to fit
def func(x,a):
    return (-(a*3000*1.347*(10**(-4))*(10**(-7))*9.8*x*(10**(-3)))/(0.0105/2))*((((0.00597/2)**2)+(((-0.115)+(1*9.8*((x*(10**(-3)))**2)/2)+(0.0105/2))**2))**((-3)/2))-((((0.00597/2)**2)+(((-0.115)+(1*9.8*((x*(10**(-3)))**2)/2)-(0.0105/2))**2))**((-3)/2))

# Generate some data based on the function
data = np.loadtxt("/content/data2.txt")
x = data[:,0]
y = data[:,1]

params, params_covariance = curve_fit(func, x, y)
print(params)
yf = func(x,(13)*(-6889.38553163))
print("Error in the estimation of the magnetic moment of the magnet is",13*np.sqrt(np.diag(params_covariance))[0])
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='Data')
plt.plot(x, yf, color='red', label='Fitted function')

plt.legend(loc='best')
plt.show()
```

```python
#This is for the SEE-Lab Experiment-2 where the time starts when the magnet is dropped.
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import random

# Define the form of the function you want to fit
def func(x,a):
    return (-(a*3000*1.347*(10**(-4))*(10**(-7))*9.8*x*(10**(-3)))/(0.0105/2))*((((0.00597/2)**2)+(((-0.145)+(1*9.8*((x*(10**(-3)))**2)/2)+(0.0105/2))**2))**((-3)/2))-((((0.00597/2)**2)+(((-0.145)+(1*9.8*((x*(10**(-3)))**2)/2)-(0.0105/2))**2))**((-3)/2))

# Generate some data based on the function
l=[i for i in range(0,300)]
x=np.array(l, dtype='float64')
l2=[random.randint(-167,167)*0.0001 for i in range(0,167)]
l2.extend([(-i)*2 for i in range(0,3)])
l2.extend([(-i)*2 for i in range(2,0,-1)])
l2.extend([(i)*2 for i in range(0,3)])
l2.extend([(i)*2 for i in range(2,0,-1)])
l2.extend([random.randint(-167,167)*0.0001 for i in range(177,300)])
y=np.array(l2, dtype = 'float64')

params, params_covariance = curve_fit(func, x, y)
print(params)
print("Error in the estimation of the magnetic moment of the magnet is",13*np.sqrt(np.diag(params_covariance))[0])
yf = func(x,-13*6099.71737724)

plt.figure(figsize=(6, 4))
plt.scatter(x, y, label='Data')
plt.plot(x, yf, color='red', label='Fitted function')

plt.legend(loc='best')
plt.show()
```
